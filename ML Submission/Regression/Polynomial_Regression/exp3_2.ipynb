```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
```

```python
data = pd.read_csv('Best Selling Mobile Phones.csv')
```

```python
X = data[['No']]
y = data['Sold(million)']
```

```python
linear_model = LinearRegression()
linear_model.fit(X, y)
```

```python
degree = 2
poly_features = PolynomialFeatures(degree=degree)
X_poly = poly_features.fit_transform(X)
poly_model = LinearRegression()
poly_model.fit(X_poly, y)
```

```python
plt.scatter(X, y, color='blue', label='Actual values')
plt.plot(X, linear_model.predict(X), color='red', label='Linear Regression line')
plt.title('Linear Regression')
plt.xlabel('No')
plt.ylabel('Sold(million)')
plt.legend()
plt.show()
```

```python
plt.scatter(X, y, color='blue', label='Actual values')
plt.scatter(X, poly_model.predict(X_poly), color='green', label='Polynomial Regression values')
plt.plot(X, poly_model.predict(X_poly), color='red', label='Polynomial Regression line')
plt.title('Polynomial Regression')
plt.xlabel('No')
plt.ylabel('Sold(million)')
plt.legend()
plt.show()
```

```python
X_high_res = np.arange(min(X['No']), max(X['No']), 0.1).reshape(-1, 1)
X_poly_high_res = poly_features.transform(X_high_res)
plt.scatter(X, y, color='blue', label='Actual values')
plt.plot(X_high_res, poly_model.predict(X_poly_high_res), color='red', label='Polynomial Regression line')
plt.title('Polynomial Regression (High Resolution)')
plt.xlabel('No')
plt.ylabel('Sold(million)')
plt.legend()
plt.show()
```

```python
new_value_linear = np.array([[8]])
predicted_value_linear = linear_model.predict(new_value_linear)
print(f'Predicted value with Linear Regression for No=8: {predicted_value_linear[0]}')
```

```python
new_value_poly = poly_features.transform(np.array([[8]]))
predicted_value_poly = poly_model.predict(new_value_poly)
print(f'Predicted value with Polynomial Regression for No=8: {predicted_value_poly[0]}')
```
